#### ¿*Qué onda* el fotón?

* La radiación electromagnética es, según los experimentos de difracción, una onda.
* Según cómo se absorbe y emite, en diferentes experimentos (cuerpo negro, efecto fotoeléctrico, modelo de Bohr) una partícula.

##### ¿Cuál de éstas es cierta?
Ambas. A esto lo llamamos ***complementariedad***: ambos modelos deben convivir, y son como la media naranja *le une para le otre*.

#### La luz, algunas particularidades.

* La luz viaja a la velocidad de ***ella misma***, $c = \lambda f$.
* No tiene masa en reposo, $m_0 = 0$ (relatividad especial).

También:

* Cuando pasa por dos rendijas, tenemos patrones de interferencia.
* Cuando intercambia energía, lo hace de manera instantánea y por cuantos de energía $E = hf$.

Particularmente, $$\lambda f = c \Longrightarrow \lambda \frac{E}{h} = c \Longrightarrow$$ $$\boxed{\Longrightarrow E = \frac{hc}{\lambda}}$$

* Al aumentar la energía disminuye la longitud de onda.

#### Las partículas con onda de De Broglie.

De Broglie propuso que:

* Las partículas con masa $m$ también poseen propiedades ondulatorias, así como las ondas EM tienen propiedades de partícula.
* La longitud de onda de la onda guía de la partícula es $\lambda = \frac{h}{p}$

La materia, bajo ciertas condiciones, tendrá propiedades ondulatorias, afirmó De Broglie.

### Las partículas con onda de De Broglie.
El respaldo fuerte de esta teoría, es que puede explicar las órbitas del modelo de átomo de Bohr diciendo que el ***radio total*** de una órbita tiene que ser un **múltiplo entero de la longitud de onda** de De Broglie.

## Davisson-Germer y difracción de electrones

* en 1927 (hace 99 años) Davisson y Germer, de manera accidental, consiguen medir la difracción de electrones en un cristal de Níquel.

* La difracción es la creación de máximos y mínimos de intensidad para diferentes ángulos, lo que se conoce como patrón de difracción. $$2d \;sin(  \theta)  = n\lambda$$

donde la distancia $d$ es la separación entre centros de dispersión y $\theta$ es el ángulo de medida, siendo $\lambda$ la longitud de la onda incidente (en este caso los electrones).

* Siendo que la distancia entre los átomos de níquel era conocida (difracción de rayos X de $\lambda$ conocida), $d = 2.15 \mathring{\mathrm{A}}$, y la longitud de onda de los electrones puede variarse median un potencial, pudieron levantar la curva $\lambda = \frac{h}{mv}$...
