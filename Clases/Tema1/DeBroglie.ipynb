### Una sola onda no alcanza -  Partícula Cuántica

Las partículas y las ondas difieren ***demasiado***, si las pensamos desde el espacio:

* Una partícula es ***puntual***, es decir, no tiene dimensiones y está perfectamente ***localizada*** en el espacio.

* Una onda senoidal simple es infinitamente larga (no hay restricciones para el dominio $x$), con lo que está ***completamente deslocalizada*** en el espacio.

### Una sola onda no alcanza -  Partícula Cuántica

**Nota**: Una onda senoidal, que se propaga *espaciotemporalmente*, con longitud de onda $\lambda$ y frecuencia $f = 1/\tau$ puede escribirse como:
$$y(x,t) = A cos\Bigg( \; \overbrace{\Bigg(\frac{2\pi}{\lambda}\Bigg)}^{\kappa = 2\pi /\lambda} x \; - \; \overbrace{\Bigg( \frac{2\pi}{\tau}\Bigg)}^{\omega = 2\pi f = 2\pi / \tau} t \Bigg) \Longrightarrow$$

$$\Longrightarrow \boxed{y(x,t) = A\; cos \Big( \kappa x - \omega t \Big) }$$

* $A$ es la amplitud de la onda.
* $\kappa [rad/m] = 2\pi /\lambda$, es el **número de onda**.
* $\omega [rad/s] = 2\pi f$, es la **frecuencia angular**.

### Una sola onda no alcanza -  Partícula Cuántica
* Se puede lograr algo ***parecido a la localización*** de una partícula mediante la suma de ondas de distinta longitud de onda. En efecto:
$$y_1(x,t) = A \; cos(\kappa_1 x - \omega_1 t) \quad y_2(x,t) = A \; cos(\kappa_2 x - \omega_2 t)$$

$$y = y_1 + y_2 = 2A\; \Big(cos(\kappa_1 x - \omega_1 t) +  cos(\kappa_2 x - \omega_2 t)\Big)$$

Esta id. trigonométrica va a servir: $ cos(a) + cos(b) = 2 cos\Big(\frac{a - b}{2}\Big) cos\Big(\frac{a + b}{2}\Big)$, entonces:

$$y(x,t) = 2A cos\Big( \Delta \kappa \,x - \Delta\omega \, t \Big) \; cos\Big(\bar{\kappa} \,x - \bar{\omega} \, t\Big)$$
con:
$$\Delta \kappa = (\kappa_1- \kappa_2)/2 \quad \Delta\omega = (\omega_1 - \omega_2)/2$$
y
$$\bar{ \kappa} = (\kappa_1+ \kappa_2)/2 \quad y \quad\bar{\omega} = (\omega_1 +   \omega_2)/2$$



```julia
x = -10:0.05:10
λ1 = 1
λ2 = 1.15
Δk = 2π*(1/λ2 - 1 / λ1) 
y1 = cos.( (2π/λ1) .* x)
y2 = cos.( (2π/λ2) .* x)
ysum = (y1 + y2) /2
yenv = cos.(Δk / 2 .* x);
```

```julia
using Measures, LaTeXStrings
g = plot(x,y1, size=(1100, 600), linewidth =2, xticks = 0, yticks = 0, label = L"$y_1$", legend = (0.2,0.915),
    title = "Sumando dos longitudes de onda parecidas, obtenemos paquetes.",
    legendfontsize = 14, alpha = 0.3, legend_columns = -1, ylims = (-1, 1.4))
plot!(x,y2, linewidth =2, label = L"$y_2$", alpha = 0.3)
plot!(x,ysum, linewidth =3, color = :black, label = L"$\frac{y_1 + y_2}{2}$");
plot!(x,yenv, linewidth = 3, line = :dash, color = :gray, label = L"$y_{envolvente}$");

hline!([0], lwd = 1.5, color = :black,linestyle = :dash, label = "")
```

### Una sola onda no alcanza -  Partícula Cuántica
#### Velocidad de Fase vs. Velocidad de grupo.

* Cada una de las ondas individuales viaja con la ***velocidad de fase***: $$v_{fase} = \lambda f = (\frac{\lambda}{2\pi}) (2\pi f) = (\frac{1}{\kappa}) (\omega) = \frac{\omega}{\kappa}$$
* La ***envonlvente*** (la cresta de cada paquete) viaja con ***velocidad de grupo***: $$v_g = \frac{\Delta \omega /2}{\Delta \kappa /2} = \frac{\Delta \omega}{\Delta \kappa}$$
* Si usamos una ***muchas*** (infinitas) longitudes de onda, tenemos que la velocidad de grupo es: $$v_g = \frac{d\omega}{d\kappa}$$

### Una sola onda no alcanza -  Partícula Cuántica
#### Velocidad de grupo y cantidades de la dinámica de partículas

Si le creemos a $\boxed{v_g = d\omega/d\kappa}$, entonces $v_g$ es la velocidad de la partícula que la onda quiere describir.
$$v_g = \frac{d\omega}{d\kappa} = \frac{\hbar d\omega}{\hbar d\kappa} = \frac{d(\hbar\omega)}{d(\hbar\kappa)}$$
* Analizando el numerador: $\hbar \omega = \frac{h}{2\pi} 2\pi f = hf = E$, la energía de la partícula.
  
* Analizando el denominador: $\hbar \kappa = \frac{h}{2\pi}\frac{2\pi}{\lambda} = \frac{h}{\lambda} = p$, el momento lineal $p = mv$ (caso bajas velocidades).

* Usando los resultados anteriores: $$\boxed{v_g = \frac{dE}{dp}}$$

### Una sola onda no alcanza - Partícula Cuántica
#### Velocidad de grupo como velocidad de la partícula.
* Caso bajas velocidades: Una partícula con momento lineal $p = m_0 v$, posee una energía $E = m_0 v^2/2 = p^2 /2 m_0$, tenemos:
$$\frac{dE}{dp} = \frac{d}{dp}\Bigg( \frac{p^2}{2 m_0}\Bigg) = \frac{2p}{2 m_0} = \frac{2 m_0 v}{2m_0} = v$$

* Caso altas velocidades: Una partícula tiene momento lineal $p = \gamma m_0 v$ posee una energía $E = \sqrt{(p c)^2 + (m_0 c^2)^2)}$, con $\gamma = \sqrt{(1-v^2/c^2)^{-1}}$,tenemos:

$$\frac{dE}{dp} = \frac{d}{dp} (p^2 c^2 + m_0 ^2 c^4)^{1/2} \overbrace{=}^{\substack{\text{saltando}\\ \text{unos} \\ \text{pasitos}} }v$$

Con esto tenemos que la velocidad de grupo describe correctamente la velocidad de la partícula, para los modelos de energía tanto newtonianos como relativistas.

### Una sola onda no alcanza -  Partícula Cuántica

* Mediante la suma de dos ondas de $\lambda 's$ similares --pero diferetes-- obtenemos ***localización***, que es una de las propiedades de una partícula, aunque la cantidad de "paquetes" vuelve a ser infinita (hacen falta muchas longitudes de onda (una distribución, de hecho).

* Se pueden sumar ***muchas*** longitudes de onda, sacadas de una distribución normal.

```julia
#Función para ver paquetes de onda.
using Distributions
function paketenga(λ_media::Float64, σ::Float64, n::Int64)
    d = Normal(λ_media, σ);
    λ = rand(d, n);
    #reemplazar esto por una dist gaussiana.
    #λ = range(λ_min,λ_max, length = n)
    x = -60:0.01:60
    y = zeros(length(x))
    
    for i ∈ λ
    y = y + cos.( (2π/i) .* x)
    end
    
    y = y / length(λ)
    return DataFrame(x = x, y = y)
end
```

```julia
λ = 4.0; #longitud de onda central
σ1 = 0.1; σ2 = 0.25; σ3 = 1.0; #desviaciones estándar
```

```julia
d1 = Normal(λ, σ1)
d2 = Normal(λ, σ2)
d3 = Normal(λ, σ3)

gdist = plot(x-> pdf(d1, x), xlim = (λ - 4*σ3,λ + 4*σ3), color = :black, legendfontsize = 18, linewidth = 4,
    title = "En lugar de dos, ponemos una distribución de funciones de onda",
    xlabel = "λ (m) \n ----------------------------------------------------------------------",
    ylabel = L"Densidad de Prob. $(m^{-1})$",
    label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ1, L"$\, )\; m$" ))
plot!(x-> pdf(d2, x),color = :red,linewidth = 4,label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ2, L"$\, )\; m$" ))
plot!(x-> pdf(d3, x),color = :blue,linewidth = 4,label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ3, L"$\, )\; m$" ));
```

```julia
ondas1 = paketenga(λ,σ1,1000);
ondas2 = paketenga(λ,σ2,1000);
ondas3 = paketenga(λ,σ3,1000);

gondas = plot(ondas1.x , ondas1.y, xlabel = "x (m)",
    legend_columns = -1, ylims = (-1.1,1.5), legend = (0.125,0.9), 
    linewidth = 4, colour = :black, legendfontsize = 14,
    title = "\n Sumando muchas λ's obtenemos solo un paquete",
    label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ1, L"$\, )\; m$" ))
plot!(ondas2.x , ondas2.y , linewidth = 4, colour = :red, label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ2, L"$\, )\; m$" ))
plot!(ondas3.x, ondas3.y , linewidth = 4, colour = :blue, label = string(L"$ \lambda \pm \sigma_\lambda = (\,$", λ, L"$\, \pm \,$", σ3, L"$\, )\; m$" ));
```

```julia
plot(gdist, gondas, layout = (2,1), left_margin = [10mm 0mm])
```

### Una sola onda no alcanza -  Partícula Cuántica

* Al sumar muchas ondas, tenemos una partícula ***localizada*** (no completamente), que viaja con una distribución de longitudes de onda --> viaja con una distribución de momentos --> viaja con una distribución de velocidades (no es una única velocidad porque tenemos muchas $\lambda 's$.

* Es importante notar que a medida que $\lambda$ crece, los momentos $p$ decrecen

* ...**PERO** si **crece** la desviación estándar $\mathbf{\sigma_lambda}$, entonces **crece** la desviación estándar de los momentos $\mathbf{\sigma_p}$: mayores dispersiones en la longitud de onda hacen también a momentos más dispersos.

* Un chiste no menor es observar el **tamaño del paquete de ondas** pensando en las **dispersiones**. Mirar Fijo el gráfico anterior y observar la relación...

### Una sola onda no alcanza -  Partícula Cuántica
* Vimos que para describir la **localización** de una partícula, es necesario usar muchas longitudes de onda (paquete gaussiano).

* Al hacerlo, definimos **localización**, pero hacemos más disperso el **momento lineal**, o la velocidad (dado que newtonianamente la masa es constante, aunque también relativísticamente funciona dE/dpen).

### El Principio de Incertidumbre de Heisenberg

El principio de incertidumbre establece que las desviaciones estándar del momento lineal $\sigma_{p_x}$ y la posición $\sigma_x$ (llamadas incertidumbres de medición en el libro) en un estado cuántico cualquiera, cumplen una desigualdad, llamada principio de incertidumbre:

$$\sigma_{p_x} \sigma_x \geq \frac{\hbar}{2} \text{\quad que es lo mismo que \quad} \Delta x \, \Delta p_x \geq \frac{\hbar}{2} $$

esto es, el principio de incertidumbre tiene que ver con una restricción a la construcción de estados cuánticos. Es presentada así en el libro por varios motivos, aunque principalmente la idea de que el principio de incertidumbre está asociado a mediciones se introduce por la falta de uno de los axiomas de la MC, el conocido, critacado, discutido ***axioma de la medida***.

Nada tiene que ver con una medida, ni con la precisión de esta (de hecho, para ***medir*** la desigualdad, hacen falta una colección de medidas...bastantes medidas).


### El principio de incertidumbre de Heisenberg.

Una manera de interpretarlo:
* Para obtener una partícula cuántica ***localizada*** debemos sumar muchas ondas con una distribución de $\lambda's$, donde caracterizamos el ancho de la distribución por $\sigma_\lambda$.

* Como vimos, si $\sigma_x \to 0 \text{\quad  a medida que \quad } \sigma_\lambda \to \infty $.

* Como $p = h / \lambda$, entonces $\sigma_x \to 0 \text{\quad  a medida que \quad } \sigma_p \to \infty$

### Principio de Incertidumbre: Ejemplo

La rapidez de un electrón es $v \pm \sigma_v= (5000.00 \pm 0.15)\,ms^{-1}$. Encuentre la incertidumbre mínima para la posición del electrón. 

**Solución**
Sabemos que $\sigma_x \geq \hbar / (2\sigma_p)$.

Haciendo $\sigma_p = m \sigma_v = (9.109 \times 10^{-31}\, kg) \; (0.15 ms^{-1}) = \boxed{1.36 \times 10^{-31} kg\, m \,s^{-1}}$

Luego tenemos que: 
$$
\sigma_x \geq \frac{\hbar}{2 \sigma_p} = \frac{h}{2\pi m \sigma_v }=\frac{6.62 \times 10^{-34} Js}{2\pi (9.109 \times 10^{-31} kg) \; (0.15 m s^{-1})} \simeq 0.00077 m \sim 0.77 mm
$$

Con lo que la incertidumbre en la posición va a ser, al menos, del orden de los mm.

### Principio de Incertidumbre: Ejemplo 2

Los átomos tienen niveles de energía cuantizados similares a los de los osciladores de Planck, aunque los niveles de energía de un átomo por lo general no están igualmente espaciados. Cuando un átomo hace una transición entre estados, se emite energía en la forma de un fotón.

Aunque un átomo excitado puede radiar en cualquier momento de $t = 0$ a $t \to \infty$, el intervalo de tiempo promedio después de la excitación durante el cual un átomo radia se llama vida media $\tau_{1/2}$. 

Si $\tau_{1/2} = 10^{-8} \, s$, use el principio de incertidumbre para calcular el ancho de línea $f$ producido por este tiempo de vida finito.

**Solución**

$$\Delta E \Delta t \geq \hbar /2 \longrightarrow \Delta E = \hbar / (2 \tau_{1/2} ) \simeq 1.055 \times 10^{-26} J $$

Esta incertidumbre en la energía tiene relación con la incertidumbre en la frecuencia (y la longitud de onda) del fotón emitido.

$$E = hf \Longrightarrow f = E/h \Longrightarrow \Delta f = \Delta E / h \simeq 1.59 \times 10^7 \, Hz$$

Esto puede ser pasado a longitud de onda:

$$
\lambda = c /f \Longrightarrow \Delta \lambda \overbrace{=}^{????} \frac{c}{f^2} \Delta f = \frac{\lambda^2}{c} \Delta f
$$

A medida que crece la longitud de onda, la incerteza se hace más grande...
