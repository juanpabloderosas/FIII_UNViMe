# Efecto Fotoeléctrico

Bibliografía: ***Física Universitaria con Física Moderna 2***. Sears Zemansky (Young y Freedman) (2018), Capítulo 38 (pág. 1254 - ).

#### Efecto Fotoeléctrico.

* Está basado en realizaciones experimentales entre 1877 y 1905.
* Lo interpretó ***Albertito Einstein*** y por eso le dieron el Premio Nobel de Física (1922).
* La interpretación de Einstein abarcó este fenómeno y dió coherencia con la explicación heurística de la radiación de cuerpo negro que dio Planck.

El ***efecto fotoeléctrico*** consiste en que una superficie metálica expulsa electrones al ser irradiada con ondas **EM** (estos electrones tienen el confuso nombre de ***fotoelectrones***.

* Los ***fotoelectrones*** son expulsados del material iluminado con diferentes energías cinéticas. Si pasamos de la situación **(a)** a la **(b)**, invirtiendo el potencial, tenemos un momento en el cual, al ir hacia potencials menores, ***todos los fotoelectrones son frenados por el potencial negativo entre el cátodo y el ánodo***.

* Algunos *fotoelectrones* tienen energía cinética $K_max$. Estos van a ser ***frenados*** para *algún potencial*  $V_0$, que llamamos ***potencial de frenado***. Usando el teorema trabajo-energía, tenemos:

$$
W = K_{f} - K_i = 0 - K_{max} = \overbrace{-eV_0}^{W} \Longrightarrow \\ \Longrightarrow  K_{max} = e V_0 = \frac{1}{2}m_e \, v_{max}^2 
$$

donde:
* $e = 1.602 \times 10^{-19}\;C$ es la carga del electrón
* $m_e \sim 9.109 \times 10^{-31} \, kg$


¿Cómo depende la fotocorriente a través de los electrodos...

1. ...de la frecuencia $f$ de la luz incidente?
1. ...del voltaje $V_{AC}$?
1. ...de la ***intensidad*** de luz incidente?

El modelo de ondas EM (ondas ElectroMagnéticas) asegura que las ondas transportan energía, de manera continua, cuya energía depende de la intensidad de la onda (la amplitud al cuadrado). Esto da las siguientes predicciones.


1. La frecuencia no influye en el efecto fotoeléctrico, es decir, los electrones en el material absorven ***cualquier*** frecuencia de luz, y terminan saliendo del material como ***fotoelectrones***.

1. A mayor intensidad de luz, los electrones absorven más energía, dado que $V_0 = K_{max}\; / \;e$, tenemos un potencial de frenado más alto.

1. Si suponemos que *sacar a los electrones del material* cuesta una cantidad finita de energía (función trabajo, $\phi$), entonces, para ***muy bajas intensidades de luz*** se esperaría una ***demora*** temporal dada por el tiempo que los fotoelectrones tardan en ganar la energía suficiente para *salir* del material.  

* (a) Si el potencial $\mathbf{V_{AC}\, > 0 }$ *empuja* a los ***fotoelectrones*** en la dirección  **cátodo $\longrightarrow$ ánodo**.

A medida que aumentamos el potencial, la corriente en el circuito aumenta hasta un valor estacionario.

* Si el potencial $\mathbf{V_{AC}\, < 0}$, entonces la corriente va decreciendo de a poco, hasta que llegamos a hacerla nula.

***Resultados Experimentales***

1. La fotocorriente depende de la frecuencia de la luz: para frecuencias por debajo de una frecuencia umbral, no se detecta fotocorriente. Esta frecuencia umbral es, generalmente, ultravioleta.
<br>

1. El potencial de frenado no depende de la intensidad de la luz (gráfico $i \;\;vs.\;\; V_{AC}\;$).
<br>

1. Incluso para intensidades de luz bajísimas, la fotocorriente no presenta ninguna demora.
<br>

1. Para frecuencias mayores de la luz incidente, aumenta el potencial de frenado proporcionalmente con la frecuencia.

#### Interpretación de Einstein del Efecto Fotoeléctrico.

* Einstein supuso que la luz consistía en paquetes discretos llamados ***fotones*** o *cuantos de luz*.

* Cada fotón tiene una energía que depende de su frecuencia (o color). En particular, asumiendo la constante de Planck, h, Einstein propuso:

$$
\boxed{
E = hf = \frac{h\,c}{\lambda}
}
$$

siendo $h = 6.62607015\; \times \; 10^{-34} \; J \, s = 4.135667696 \; \times \; 10^{-15} \; eV \, s$ la constante que Planck había asumido para el cálculo de la radiación de cuerpo negro.



**Interpretación de Einstein del Efecto Fotoeléctrico**

* Los electrones del material pueden:

    * No captar nada de energía de un fotón.

    * Captar **toda** la energía de un fotón.

***Resultados Experimentales***: La fotocorriente depende de la frecuencia de la luz: para frecuencias por debajo de una frecuencia umbral, no se detecta fotocorriente. Esta frecuencia umbral es, generalmente, ultravioleta.

Einstein: Si ***sacar a un electrón de un material*** cuesta ***la función trabajo*** $\phi$, entonces:

* Si la energía del fotón es mayor que la función trabajo, $hf > \phi$, se emiten fotoelectrones.

* Si la energía es menor, $\phi > hf$, la energía de los fotones no alcanza...y no hay fotoelectrones.

***Resultados Experimentales***: 2. El potencial de frenado no depende de la intensidad de la luz (gráfico $i \;\;vs.\;\; V_{AC}\;$).

**Einstein**: Depende de la frecuencia:

$$
K_{max} = E - \phi = hf - \phi  \Longrightarrow \\
\; \\ \;
\boxed{
V_0 = \frac{h}{e} f - \frac{\phi}{e} 
}
$$

***Resultados Experimentales***: ***Resultados Experimentales***: Incluso para intensidades de luz bajísimas, la fotocorriente no presenta ninguna demora.


**Einstein**: No hay proceso ***continuo***, sino ***discreto*** de absorción de energía: el electrón absorbe un fotón y gana energía (y al contrario).

Entonces, no puede haber demora: la energía se transmite a ***todo o nada***.
